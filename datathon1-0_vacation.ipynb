```python
train_data=pd.read_csv("/kaggle/input/gdsc-pesu-rr-datathon/train.csv")
train_data
```

```python
#making vaction rate in correct range ie 1-100 in test data
indexAge = train_data[ train_data['VACATION_RATE']>100].index
train_data.drop(indexAge , inplace=True)
max(train_data['VACATION_RATE'])
```

```python
#making the T2M_MAX in celcius
#train_data['T2M_MAX']-train_data['T2M_MIN']-273.1523  #this equals train_data['T2M_RANGE']
train_data['T2M_RANGE']=train_data['T2M_MAX']-train_data['T2M_MIN']
```

```python
train_data.head()
```

```python
test_data=pd.read_csv("/kaggle/input/gdsc-pesu-rr-datathon/test.csv")
test_data
```

```python
#Dropping variables not in test
train_data=train_data.drop(['T2M','QV2M'],axis=1)
test_data=test_data.drop('ID',axis=1)
train_data.head()
```

```python
#trying linear regression (submission 1! :)
X=train_data.drop('VACATION_RATE',axis=1)
Y=train_data['VACATION_RATE']
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

from sklearn.linear_model import LinearRegression
lr = LinearRegression().fit(X_train,Y_train)
lr.coef_
```

```python
Y_pred=lr.predict(X_test)
len(X_test)
Y_pred
```

```python
df=pd.read_csv('/kaggle/input/gdsc-pesu-rr-datathon/test.csv')
X1=df.drop('ID',axis=1)
X1['T2M_RANGE']=X1['T2M_MAX']-X1['T2M_MIN']
Y_pred=lr.predict(X1)
print(Y_pred)
```

```python
df['VACATION_RATE']=Y_pred
df[['ID','VACATION_RATE']].to_csv('submission.csv',index=False)
```

```python
#trying with xgboost
from xgboost import XGBRegressor
xgb_reg = XGBRegressor(random_state=42)
xg=xgb_reg.fit(X_train, Y_train)
Y_pred=xg.predict(X_test)
```

```python
df=pd.read_csv('/kaggle/input/gdsc-pesu-rr-datathon/test.csv')
X1=df.drop('ID',axis=1)
X1['T2M_RANGE']=X1['T2M_MAX']-X1['T2M_MIN']
Y_pred=xg.predict(X1)
print(Y_pred)
df['VACATION_RATE']=Y_pred
df[['ID','VACATION_RATE']].to_csv('submission.csv',index=False)
```
